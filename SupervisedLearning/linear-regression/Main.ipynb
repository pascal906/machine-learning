```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

# Generate Data

```python
jumlah_data = 100

y = np.array([i*0.1+np.random.randn() for i in range(jumlah_data)])
x = np.array([i*0.1 for i in range(jumlah_data)])
```

## 1. Plot Data

```python
plt.scatter(x, y)
plt.show()
```

## 2. Fungsi Linear

```python
def fungsi_linear(x, m):
    y = m * x
    return y
```

### Menghitung persamaan Linear Awal

```python
x_prediksi = np.array([0, 10])
m_awal = 5
y_prediksi = fungsi_linear(x_prediksi, m_awal)
```

```python
# print(x_prediksi)
# print(y_prediksi)
```

```python
plt.scatter(x, y)
plt.plot(x_prediksi, y_prediksi, "Red")
plt.axis([0, 10, 0, 10])
plt.show()
```

```python
m_prediksi = m_awal
m_list_prediksi = []
y_list_prediksi = []
x_list_prediksi = []
learning_rate = 0.1 # nilai valuenya antara 0 - 1
for i in range(1, jumlah_data):
    y_prediksi = fungsi_linear(x[i], m_prediksi)
    y_aktual = y[i]

    error = y_aktual - y_prediksi

    delta_m = learning_rate * error / x[i]

    m_prediksi = m_prediksi + delta_m

    m_list_prediksi.append(m_prediksi)
    # y_list_prediksi.append(y_prediksi)
    # x_list_prediksi.append(x[i])
    y_list_prediksi.append(fungsi_linear(np.array([0, 10]), m_prediksi))
    x_list_prediksi.append(np.array([0, 10])) 
```

## Visualisasi Prediksi

```python
fig = plt.figure(figsize = (4,4))
line, = plt.plot([],[], "Red")

def animate(frame_num):
    x = x_list_prediksi[frame_num]
    y = y_list_prediksi[frame_num]
    line.set_data((x, y))
    return line

plt.scatter(x, y)
plt.axis([0, 10, 0, 10])
pre_visual = FuncAnimation(fig, animate, frames = 100, interval = 100, repeat = False)
plt.show()
```
